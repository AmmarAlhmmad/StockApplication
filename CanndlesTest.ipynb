```python
import pandas as pd
import math
df = pd.read_excel('Raghi2018.xlsx')
df['type']
# initiate class
class Invest : 
    def __init__(self, day , close , openo , high, low, typeo):
        self.day = day
        self.close = close 
        self.openo = openo 
        self.high = high 
        self.low = low
        self.typeo = typeo 
# iterate over rows and store each row's attribute as invest object
rowslist = df.values.tolist()
invest_instances = []
for i in rowslist:
    invest_instances.append(Invest(*i))

    
    
buy_Notification = 0
buy_Confirmation = 0
sell_Notification = 0
sell_Confirmation = 0




for indx,j in enumerate(invest_instances): #skiping first row
    if (indx>=1) : 
        if (invest_instances[indx].typeo > 0 and invest_instances[indx-1].typeo < 0) : #buy case: green and the one before it red
            if (invest_instances[indx].close > invest_instances[indx-1].openo) :       #close for the green is larger than the open of the red before it 
                buy_Notification+=1
                print("for index:"+ str(indx) + "::::")
                print("green close is: ")
                print(invest_instances[indx].close)
                
                 
                print("and red open is: ")
                print(invest_instances[indx-1].close)
                print("hence let's buy!")

                if (indx< len(invest_instances) -1) :
                    #print((invest_instances[indx+1].typeo))
                    if(invest_instances[indx+1].typeo > 0): 
                        print("for index: " +str(indx+1))
                        print("the third one is green!")
                        #print((invest_instances[indx+1].close))
                        
                        buy_Confirmation+=1
                    
                    
        elif(invest_instances[indx].typeo < 0 and invest_instances[indx-1].typeo > 0):
            if(invest_instances[indx].openo > invest_instances[indx-1].close):    #open for red is bigger than the close for the green before it 
                sell_Notification +=1 
                print("for index:"+ str(indx) + "::::")
                print("green close is: ")
                print(invest_instances[indx-1].close)
                
                 
                print("and red open is: ")
                print(invest_instances[indx].close)
                print("hence let's sell!")
                
                if (indx< len(invest_instances) -1) :
                    if(invest_instances[indx+1].typeo < 0):
                        sell_Confirmation+=1
                        print("for index: " +str(indx+1))
                        print("the third one is red!")

        
        
        
        
        
    #print(indx,j)

print("buy notification: "+ str(buy_Notification))
print("buy confirmation:" + str(buy_Confirmation))
buy_Accuracy=round(buy_Confirmation/buy_Notification*100)
print("buy accuracy:" + str(buy_Accuracy)+"%")
print("sell notification: "+ str(sell_Notification))
print("sell confirmation:" + str(sell_Confirmation))
sell_Accuracy=round(sell_Confirmation/sell_Notification*100)
print("sell accuracy:" + str(sell_Accuracy) + "%")
```
